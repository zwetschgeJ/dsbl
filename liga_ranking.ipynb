```python
def get_data_google(link_id: str = '1LeFExyFaH101326yN82BKjLaPmX0uKKTK4g0lQvBquY') -> pd.DataFrame:
    df = pd.read_csv(
        'https://docs.google.com/spreadsheets/d/' +
        link_id  +
        '/export?gid=0&format=csv',
        skiprows=1,
    )
    df.drop(["1.", "Overall"], axis=1, inplace=True)
    
    columns = ["Teams", "SCP"]
    columns.extend([f'Flight {i}' for i in range(1, FLIGHTS + 1)])

    df.columns = columns
    return df
```
